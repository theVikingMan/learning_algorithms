```python
# Cost function: MSE
def compute_mse(y_pred: np.ndarray, y_train: np.ndarray) -> int: 
    n = y_pred.shape[0]
    error = y_pred - y_train
    return np.sum((1 / (2*n)) * (error**2))
```

```python
print(f'MSE for m=0.1 & b=0.1: {compute_mse(y_pred_1, y_train):.2f}')
print(f'MSE for m=9.0 & b=6.0: {compute_mse(y_pred_2, y_train):.2f}')
```

```python
# Set up gradient descent algorithm

def gradient_descent(x_train, y_train, w, b, epsilon, alpha):
    n = x_train.shape[0]
    iterations = 0
    
    y_pred = compute_linear_regression(w, b, x_train, feature)
    curr_mse = compute_mse(y_pred, y_train)
    
    while curr_mse > epsilon and iterations < 10000:
            
        w_partial = (1/n) * np.sum((x_train[feature] * (y_pred - y_train)), axis=0)
        b_partial = (1/n) * np.sum((y_pred - y_train), axis=0)
        
        w -= alpha * w_partial
        b -= alpha * b_partial
        
        y_pred = compute_linear_regression(w, b, x_train, feature)
        curr_mse = compute_mse(y_pred, y_train)
        iterations += 1
    
    return w, b, curr_mse, iterations, y_pred
```

```python
alpha, initial_w, initial_b, epsilon = 0.7, 0, 0, 0.5

optimal_w, optimal_b, final_mse, iterations, final_y_hat = gradient_descent(x_train, y_train, initial_w, initial_b, epsilon, alpha)
print(f'Final MSE: {final_mse:.2f}')
print(f'Optimal w parameter: {optimal_w:.2f}')
print(f'Optimal b parameter: {optimal_b:.2f}')
print(f'Number of iterations until convergance: {iterations:,}')
```
